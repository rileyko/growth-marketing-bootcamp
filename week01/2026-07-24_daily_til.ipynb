**2026.07.24 · 마케팅 분석**


## 오늘의 핵심

마케팅 분석은 채널이나 지표의 이름을 외우는 데서 끝나지 않는다. 고객이 해결하려는 일과 제품의 차별점을 먼저 확인한 뒤, 목적에 맞는 매체와 지표를 고르는 과정으로 이해했다.


## 배운 내용

### 매체와 과금 방식

ATL은 TV·라디오·신문·잡지처럼 불특정 다수에게 알리는 대중매체 광고다. BTL은 특정 타깃을 대상으로 하는 직접 마케팅이며, TTL은 두 방식을 함께 활용하는 통합 마케팅이다.

광고는 목표에 따라 클릭당 비용(CPC), 1,000회 노출당 비용(CPM), 전환당 비용(CPA)처럼 서로 다른 방식으로 비용을 본다. ROAS는 광고비 대비 매출을, ROI는 전체 투자 대비 순이익을 확인하는 지표다.

### 고객과 미디어를 보는 기준

Owned·Paid·Earned 미디어는 각각 자사 채널, 비용을 지불해 사용하는 매체, 리뷰·추천·입소문처럼 외부에서 획득되는 반응을 뜻한다. 앱 추적 투명성(ATT) 환경에서는 광고 플랫폼 의존도를 낮추고 퍼스트파티 데이터를 중요하게 봐야 한다는 점도 함께 확인했다.

JTBD는 고객이 해결하려는 일을 살피는 관점이다. 고객의 니즈와 페인포인트, 그 뒤에 있는 목적을 함께 확인해야 한다. USP는 이 일을 해결하기 위해 제품이 고객에게 전달할 차별화된 핵심 가치다.

### 데스크 리서치의 출발점

시장과 고객을 조사할 때는 공공 통계, 기업 공시, 산업 리포트, 광고 사례, 앱과 커뮤니티의 리뷰, 검색어와 소셜 데이터 등에서 필요한 근거를 찾을 수 있다. 검색 추이는 절대적인 수치보다 시간에 따른 변화와 성별·기기 같은 조건별 차이를 살피는 데 활용할 수 있다.


## 실습 및 적용

제품 또는 서비스의 마케팅 전략을 만들기 위한 시장조사를 시작했다. 산업과 제품을 정한 뒤 키워드, 시장 동향, 경쟁사, 고객 반응을 조사하고, 이후 JTBD와 USP 수업 내용을 더해 조사 범위를 구체화할 예정이다.

오후에는 광고 소재를 역기획하는 실습을 진행했다. 광고를 볼 때 결과물만 평가하기보다 어떤 타깃에게 무엇을 말하고, 어떤 행동을 기대하는지 거꾸로 질문해보는 방식으로 접근했다.


## 내 해석과 적용

시장조사는 AI에게 분석을 맡기는 것만으로 끝나지 않는다고 느꼈다. 직접 자료를 찾고 용어와 맥락을 확인하는 과정이 있어야 어떤 정보를 믿고 다음 질문으로 연결할지 판단할 수 있다.

마케팅에 적용할 때는 코드나 도구를 먼저 찾기보다 시장·제품·소비자를 직접 살펴보고, 그 과정에서 발견한 질문을 메시지와 광고 소재에 연결해보고 싶다.
