## 실습

문제 정의 → 데이터 확인 → 분석 → 해석 순서로 기록하세요.


```python
# 1. 잡코리아·사람인 각 100개를 수집·정규화한 결과 불러오기
import json
import pandas as pd

RAW_PATH = Path("../../outputs/marketing_jobs_200/marketing_jobs_200.json")
rows = json.loads(RAW_PATH.read_text(encoding="utf-8"))["rows"]
df = pd.DataFrame(rows).rename(columns={
    "자격요건 원문 요약": "자격 요건",
    "우대사항 원문 요약": "우대사항",
})

# 2. 인플루언서·바이럴 공고와 중복 URL 제거
df = df[~df["공고명"].str.contains("인플루언서|바이럴", na=False)]
df = df.drop_duplicates("공고 URL")

columns = ["출처", "회사명", "지역", "회사산업", "공고명", "직무군", "경력구간",
           "채용 목적", "주요업무", "자격 요건", "우대사항", "도구",
           "하드스킬", "소프트스킬", "포트폴리오 증거", "명시·추론", "공고 URL"]
df = df[columns]

# 3. 분석에 필요한 컬럼만 선택
assert df.groupby("출처").size().to_dict() == {"사람인": 100, "잡코리아": 100}
DATA_DIR.mkdir(parents=True, exist_ok=True)
print(f"수집 결과: {len(df)}개")
df.head()
```

```python
# 필수 기준에 맞게 JD 요약표를 한 번 더 정리
import re
from IPython.display import display

jd_summary = df.copy()  # 원본 df는 그대로 보존

# 원문에 자격요건·우대사항·도구가 하나라도 있으면 JD 명시로 구분
detail_cols = ["자격 요건", "우대사항", "도구"]
has_jd_detail = ~jd_summary[detail_cols].fillna("").apply(
    lambda col: col.str.strip().isin(["", "기재 없음"])
).all(axis=1)

# 1. 없는 내용을 직무별 템플릿으로 채우지 않고 그대로 구분
missing_label = "JD에서 확인되지 않음"
for col in ["자격 요건", "우대사항", "도구"]:
    missing = jd_summary[col].fillna("").str.strip().isin(["", "기재 없음"])
    jd_summary.loc[missing, col] = missing_label

# 2. 다중 선택 항목의 구분자를 '/'로 통일하고 중복 키워드 제거
def clean_items(value, limit=7):
    items = [x.strip() for x in re.split(r"\s+[·/|]\s+", str(value)) if x.strip()]
    return " / ".join(dict.fromkeys(items[:limit]))

for col in ["주요업무", "도구", "하드스킬", "소프트스킬", "포트폴리오 증거"]:
    jd_summary[col] = jd_summary[col].map(clean_items)
    jd_summary[col] = jd_summary[col].replace("기재 없음", missing_label)

portfolio_is_explicit = jd_summary["포트폴리오 증거"].str.startswith("포트폴리오 제출")
jd_summary["포트폴리오 증거"] = portfolio_is_explicit.map(
    {True: "포트폴리오 제출", False: missing_label}
)

# 3. 선택형 항목과 명시·추론 기준 정리
# 경력무관·신입/경력 공고는 최소 지원 경력을 기준으로 '신입'에 통합
jd_summary["경력구간"] = jd_summary["경력구간"].replace({
    "경력무관": "신입", "신입·경력": "신입"
})
jd_summary["명시·추론"] = has_jd_detail.map({True: "JD 명시", False: "분석자 추론"})

# 4. 필수 항목의 작성률 확인
required = ["출처", "회사명", "지역", "채용 목적", "공고명", "직무군",
            "회사산업", "경력구간", "공고 URL",
            "주요업무", "자격 요건", "우대사항", "도구", "하드스킬",
            "소프트스킬", "포트폴리오 증거", "명시·추론"]
not_ready = jd_summary[required].apply(
    lambda col: col.fillna("").astype(str).str.strip().isin(["", "기재 없음", "확인 필요", missing_label])
)
quality = pd.DataFrame({
    "필수 항목": required,
    "작성 완료": (~not_ready).sum().values,
    "확인 필요": not_ready.sum().values,
})
quality["작성률"] = (quality["작성 완료"] / len(jd_summary) * 100).round(1).astype(str) + "%"

print(f"필수 항목 정리 완료: {len(jd_summary)}개")
display(quality.style.hide(axis="index"))
display(jd_summary.head(10).style.hide(axis="index"))
```

```python
# 전체 공고와 신입 대상 공고를 같은 컬럼 구조로 분리
all_jobs = jd_summary.copy()
entry_jobs = all_jobs[all_jobs["경력구간"] == "신입"].copy()

# 직무별로 신입 공고에서 자주 등장한 요구사항 요약
def top_items(series, n=5):
    values = series.dropna().str.replace("A/B 테스트", "A-B 테스트", regex=False)
    items = values.str.split("/").explode().str.strip()
    items = items[items != missing_label]
    return " / ".join(items.value_counts().head(n).index) or "확인된 항목 없음"

def top_values(series, n=3):
    return " / ".join(series.value_counts().head(n).index)

entry_role_summary = (
    entry_jobs.groupby("직무군", observed=True)
    .agg(
        공고수=("공고 URL", "count"),
        주요산업=("회사산업", top_values),
        주요업무=("주요업무", top_items),
        주요도구=("도구", top_items),
        하드스킬=("하드스킬", top_items),
        소프트스킬=("소프트스킬", top_items),
        포트폴리오_증거=("포트폴리오 증거", top_items),
    )
    .reset_index()
    .sort_values("공고수", ascending=False)
)

ENTRY_SUMMARY_PATH = DATA_DIR / "marketing_jobs_entry_role_summary.csv"
entry_role_summary.to_csv(ENTRY_SUMMARY_PATH, index=False, encoding="utf-8-sig")

# 두 데이터셋은 행 개수만 다르고 컬럼은 완전히 동일
assert all_jobs.columns.equals(entry_jobs.columns)
assert entry_jobs["경력구간"].eq("신입").all()
print(f"전체: {len(all_jobs)}개 / 신입 대상: {len(entry_jobs)}개")
display(entry_role_summary.style.hide(axis="index"))
display(entry_jobs.head(10).style.hide(axis="index"))
```

```python
# 과제에서 지정한 12개 속성만 정확한 순서로 제출용 데이터 생성
assignment_columns = [
    "채용 목적", "공고명", "직무군", "회사·산업", "경력구간", "공고 URL",
    "주요업무", "도구", "하드스킬", "소프트스킬", "포트폴리오 증거", "명시·추론",
]

assignment_all = (
    all_jobs.rename(columns={"회사산업": "회사·산업"})[assignment_columns]
    .copy()
)
assignment_entry = assignment_all[assignment_all["경력구간"] == "신입"].copy()

ASSIGNMENT_ALL_PATH = DATA_DIR / "jd_summary_assignment_all.csv"
ASSIGNMENT_ENTRY_PATH = DATA_DIR / "jd_summary_assignment_entry.csv"
assignment_all.to_csv(ASSIGNMENT_ALL_PATH, index=False, encoding="utf-8-sig")
assignment_entry.to_csv(ASSIGNMENT_ENTRY_PATH, index=False, encoding="utf-8-sig")

# 과제의 컬럼 구조와 선택값만 확인
career_options = {"신입", "1~3년", "3~5년", "5년 이상"}
assert assignment_all.columns.tolist() == assignment_columns
assert set(assignment_all["경력구간"]) <= career_options
assert set(assignment_all["명시·추론"]) <= {"JD 명시", "분석자 추론"}
print(f"과제 제출용 저장 완료: 전체 {len(assignment_all)}개 / 신입 {len(assignment_entry)}개")
display(assignment_all.head(10).style.hide(axis="index"))
```

```python
# 능력 분류표: 전체 200개 공고에서 공통 역량과 직무별 특징 비교
ability_parts = []
for col, prefix in [("하드스킬", ""), ("소프트스킬", ""), ("도구", "도구: ")]:
    part = assignment_all[["공고 URL", "직무군", col]].copy()
    part["능력"] = part[col].fillna("").str.split(" / ")
    part = part.explode("능력")
    part["능력"] = prefix + part["능력"].str.strip()
    part = part[~part["능력"].str.contains(missing_label, na=False)]
    ability_parts.append(part[["공고 URL", "직무군", "능력"]])

ability_long = pd.concat(ability_parts).drop_duplicates()
role_sizes = assignment_all["직무군"].value_counts()
ability_counts = ability_long.groupby(["직무군", "능력"])["공고 URL"].nunique()
denominator = pd.Series(
    ability_counts.index.get_level_values(0).map(role_sizes).to_numpy(),
    index=ability_counts.index,
)
ability_rates = ability_counts.div(denominator).unstack(0, fill_value=0).fillna(0)

# 네 직무 모두에 등장한 일반 역량 중 상위 항목을 공통 능력으로 선정
domain_words = r"CRM|광고|콘텐츠|퍼널|SEO|SNS|이메일|카피|디자인|영상|도구:"
common_pool = ability_rates[(ability_rates > 0).all(axis=1)]
common_pool = common_pool[~common_pool.index.str.contains(domain_words, regex=True)]
common_skills = common_pool.mean(axis=1).nlargest(6).index.tolist()

# 직무별 항목은 다른 직무보다 상대적으로 더 자주 나온 능력을 선택
def distinctive_skills(role, n=6):
    other_roles = ability_rates.columns.drop(role)
    lift = ability_rates[role] - ability_rates[other_roles].mean(axis=1)
    candidates = lift[(ability_rates[role] > 0) & ~lift.index.isin(common_pool.index)]
    return candidates.nlargest(n).index.tolist()

ability_table = pd.DataFrame([
    ["🟢 공통 능력 (모든 마케터)", " / ".join(common_skills)],
    ["🔵 CRM 마케터", " / ".join(distinctive_skills("CRM"))],
    ["🔵 퍼포먼스 마케터", " / ".join(distinctive_skills("퍼포먼스"))],
    ["🔵 콘텐츠 마케터", " / ".join(distinctive_skills("콘텐츠"))],
    ["🔵 그로스 마케터", " / ".join(distinctive_skills("그로스"))],
], columns=["구분", "능력 / 역량"])

display(
    ability_table.style.hide(axis="index")
    .set_properties(**{"text-align": "left", "white-space": "normal"})
    .set_table_styles([{"selector": "th", "props": [("text-align", "left")]}])
)
```
