**2026.07.23 · 오늘의 학습 기록**


## 오늘의 핵심

마케팅에서 리서치와 데이터 분석이 필요한 이유는 자료를 많이 모으기 위해서가 아니라, 고객의 선택과 이탈을 더 정확한 질문으로 바꾸기 위해서다. AI와 데이터는 답을 대신 내리는 주인공이 아니라, 마케터가 세운 방향을 확인하고 실행 범위를 넓히는 도구로 사용해야 한다.


## 학습 기록

### 목적에서 시작하는 마케팅 리서치

리서치를 시작할 때 먼저 물어야 할 것은 ‘무엇을 조사할까?’보다 ‘어떤 판단을 내리기 위해 조사하는가?’였다. 신제품 출시, 매출 변화 원인 확인, 트렌드 분석은 필요한 근거가 서로 다르다. 정해진 분석 틀을 먼저 고르는 대신 목적에 맞춰 시장·경쟁사·고객 중 어디를 깊게 볼지 결정해야 한다.

시장 규모만 확인해서는 고객이 브랜드를 어떻게 경험하는지 알기 어렵다. 고객의 현재 행동과 맥락, 드러난 니즈와 잠재 니즈, 불편을 함께 살펴야 제품과 시장이 맞는지 판단할 수 있다. 고객에게 원하는 답을 직접 묻는 데 그치지 않고 실제 선택과 행동을 관찰해야 한다는 점도 중요했다.

PMF를 확인할 때는 제품을 출시했다는 사실보다 고객이 계속 사용하는지, 추천할 만큼 만족하는지를 봐야 한다. MVP는 큰 비용을 들이기 전에 핵심 가정을 작게 검증하는 수단으로 이해했다. 결국 리서치는 새로운 말을 만들어내는 작업이 아니라, 이미 내린 판단에서 한 걸음 떨어져 시야를 넓히는 과정에 가깝다.

### 고객 이탈을 질문으로 바꾸는 그로스 마케팅

고객 여정은 기획·생산·판매로 끝나는 일방향 과정이 아니었다. 고객 반응을 확인하고 제품과 메시지를 고친 뒤 다시 배포하는 순환 구조에 가깝다. 그래서 AARRR은 지표를 채우는 표보다 고객이 어느 단계에서 왜 멈추는지 찾는 질문의 순서로 보는 편이 유용하다.

획득 이후 활성화가 낮다면 타깃이나 메시지가 실제 경험과 어긋났는지, 진입 과정에 마찰이 있는지 확인할 수 있다. 유지 단계에서는 고객이 핵심 가치를 경험했는지, 수익화 단계에서는 가격과 신뢰가 선택을 막는지 살펴볼 수 있다. 전환율만 높이려 하기보다 이탈을 자연스러운 전제로 두고 원인을 좁혀야 다음 행동이 구체적으로 나온다.

이 과정에서 통계는 눈에 띄는 결과를 곧바로 원인으로 단정하지 않게 도와준다. 표본이 충분한지, 우연한 변동인지, 살아남은 결과만 보고 있지는 않은지 확인해야 한다. 데이터 분석의 역할은 멋진 대시보드를 만드는 것이 아니라 문제 정의와 마케팅 행동 사이의 근거를 단단하게 만드는 데 있다.

### AI와 데이터는 판단을 보조하는 도구

AI는 자료를 정리하고 가설이나 콘텐츠 방향을 넓히는 데 유용하지만, 무엇을 물을지와 어떤 결과를 사용할지는 마케터가 결정해야 한다. 마케터가 본질적인 질문과 방향을 세우고, AI가 실행 범위를 넓힌 뒤, 다시 사람이 브랜드의 맥락에 맞게 고르는 흐름이 필요하다.

나는 데이터와 AI를 다뤄온 경험이 강점이지만, 그 강점만 앞세우면 기술을 구현하는 사람으로 보일 수 있다는 고민이 있다. 앞으로는 분석 자체보다 소비자의 반응에서 무엇을 발견했고, 메시지와 콘텐츠를 어떻게 바꿨는지를 보여줘야 한다. 기술을 숨기기보다 마케팅 판단을 뒷받침하는 근거로 배치하는 것이 내가 만들고 싶은 방향이다.


## 실습 및 적용

### 나의 언어로 마케팅 정의하기

오늘은 ‘마케팅이 무엇인지’, ‘마케팅을 잘한다는 것은 무엇인지’, ‘AI와 데이터를 어떻게 활용할지’를 내 경험과 연결해 정리했다.

내가 생각하는 마케팅은 사람들이 무엇을 선택하고 망설이는지 관찰하고, ‘왜 그럴까?’라는 질문에서 이유를 찾아가는 일이다. 찾은 이유를 바탕으로 브랜드가 다음에 무엇을 말하고 보여줄지 결정하는 과정까지 포함한다.

마케팅을 잘한다는 것은 소비자 반응에서 발견한 내용을 설명하는 데서 끝내지 않고 실제 메시지와 콘텐츠에 반영하는 것이다. 선택을 방해하는 요소는 줄이고 필요한 이유는 더해, 소비자의 실제 행동으로 이어지게 만들어야 한다.

AI와 데이터 분석은 이 판단을 빠르게 검토하는 데 사용한다. 데이터로 반복되는 선택과 반응을 확인하고, AI로 자료 정리와 가설 탐색의 범위를 넓힐 수 있다. 다만 결과를 그대로 따르지 않고 소비자의 맥락과 실제 반응을 함께 보며 실행할 것과 버릴 것을 결정해야 한다.


## 배운 점과 다음 액션

오늘 가장 크게 달라진 생각은 리서치와 데이터 분석의 출발점이 도구나 정해진 형식이 아니라는 점이다. 먼저 마케팅 목적과 고객에 관한 질문을 정해야 필요한 자료와 지표를 고를 수 있다. 콘텐츠 마케터도 단순한 제작자가 아니라 고객을 이해하고 기획부터 실행·개선까지 연결하는 마케터라는 점을 다시 확인했다.

다음에는 AI가 정리한 답부터 보지 않고, 직접 자료를 찾아 리서치 질문과 판단 근거를 작성해보려 한다. 콘텐츠 사례를 볼 때도 예쁘거나 반응이 좋다는 평가에 그치지 않고, 어떤 고객을 겨냥했고 어느 퍼널의 마찰을 줄이려는지 분석할 것이다. 이후에는 내 데이터 경험을 B2C 콘텐츠 마케팅 사례로 번역해 포트폴리오에서 보여줄 근거를 정리할 계획이다.

아직 확인이 필요한 것은 PMF 이후의 성장률을 어떤 지표와 기간으로 판단할지, 목적에 따라 바뀌더라도 공통으로 확인해야 하는 최소 리서치 항목이 무엇인지다. 퍼널 단계별 지표와 아하 모먼트, 현재 집중할 단일 지표를 실제 사례에 적용하면서 기준을 구체화해야 한다.
